# Pydantic

```python
class CleanEvent(BaseModel):
    event_id: str
    name: str
    date: Optional[date] = None
    time: Optional[time] = None
    region: Optional[str] = None
    city: Optional[str] = None
    coordx: Optional[float] = None
    coordy: Optional[float] = None
    venue: Optional[str] = None
    segment: str
    genre: Optional[str] = None
    sub_genre: Optional[str] = None

    @field_validator("event_date", mode="before")
    @classmethod
    def parse_date(cls, v):
        if not v:
            return None
        # Ticketmaster sometimes gives date-only, sometimes full datetime
        return v[:10] if isinstance(v, str) else v
```

ID
Name - str
Date - date
Time - time
Location  - str/str/str/(float,float)
Coordinate
Region - categorical 
Segment - categorical
Genre (optional) - categorical


Features - Columns:

- Name (required)
- Timestamp (required) - date (required), time (optional)
- Location (required) - region (required), city (required), coords (optional), venue (optional)
- Type (required) - Segment (required), Genre (optional), Sub_genre (optional)

Convert segment, genre, region to categories in the analysis stage (pandas)

```python
def flatten_event(raw: dict) -> dict:
    classifications = raw.get("classifications", [{}])[0]
    venues = raw.get("_embedded", {}).get("venues", [{}])
    venue = venues[0] if venues else {}
    location = venue.get("location", {})

    dates = raw.get("dates", {}).get("start", {})

    return {
        "event_id": raw.get("id"),
        "name": raw.get("name"),
        "date": dates.get("localDate"),
        "time": dates.get("localTime"),
        "region": venue.get("state", {}).get("name"),
        "city": venue.get("city", {}).get("name"),
        "coordx": location.get("longitude"),
        "coordy": location.get("latitude"),
        "venue": venue.get("name"),
        "segment": classifications.get("segment", {}).get("name"),
        "genre": classifications.get("genre", {}).get("name"),
        "sub_genre": classifications.get("subGenre", {}).get("name"),
    }
```
